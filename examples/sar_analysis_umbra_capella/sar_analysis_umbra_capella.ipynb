# Umbra and Capella SAR Analysis with OpenGeo

This notebook demonstrates how to access **Umbra** and **Capella** Synthetic Aperture Radar (SAR) data.

## 1. Initialize Umbra

```python
import opengeo as og

# Initialize Umbra STAC (Canopy V2)
og.Initialize("UMBRA")
```

## 2. Load Umbra SAR Collection

Note: Umbra items are often specialized. We'll search for recently created items.

```python
try:
    umbra_col = og.ImageCollection("umbra-sar-data") \
        .filterDate("2023-01-01", "2023-12-31")
    print(f"Umbra images found: {umbra_col.size()}")
except Exception as e:
    print("Could not query Umbra (check API availability or collection ID):", e)
```

## 3. Capella Initialize

```python
og.Initialize("CAPELLA")
# Search for a common Capella collection
try:
    capella_col = og.ImageCollection("capella-gec")
    print(f"Capella images found: {capella_col.size()}")
except Exception as e:
    print("Capella query failed:", e)
```
